```python
from mmdet.datasets.builder import DATASETS
from mmdet.datasets.coco import CocoDataset

@DATASETS.register_module(force=True)
class CupDataset(CocoDataset):
    CLASSES = ('polyp', )
```

```python
from mmcv import Config

config_file = '/content/mmdetection/configs/faster_rcnn/faster_rcnn_r50_caffe_fpn_mstrain_3x_coco.py'

cfg = Config.fromfile(config_file)
del cfg['data']['train']['times']
del cfg['data']['train']['dataset']
```

```python
from mmdet.apis import set_random_seed
import mmcv
import os.path as osp

cfg.dataset_type = 'CupDataset'
cfg.data_root = '/content/drive/MyDrive/대장용종 이미지 데이터/project'

#train데이터 데이터셋 정보기입
cfg.data.train.type = 'CupDataset'
cfg.data.train.data_root ='/content/drive/MyDrive/대장용종 이미지 데이터/project'
cfg.data.train.ann_file = 'potrain.json'
cfg.data.train.img_prefix = 'images/train'

#valid 데이터 데이터셋 정보기입
cfg.data.val.type = 'CupDataset'
cfg.data.val.data_root ='/content/drive/MyDrive/대장용종 이미지 데이터/project'
cfg.data.val.ann_file = 'poval.json'
cfg.data.val.img_prefix = 'images/val'

#test데이터 데이터셋 정보기입
cfg.data.test.type = '/content/drive/MyDrive/대장용종 이미지 데이터/project'
cfg.data.test.data_root ='/content/data/coco'
cfg.data.test.ann_file = 'poval.json'
cfg.data.test.img_prefix = 'images/val'

```

```python
cfg.model.roi_head.bbox_head.num_classes = 1
```

```python
cfg.load_from = '/content/mmdetection/checkpoints/faster_rcnn_r50_caffe_fpn_mstrain_3x_coco_20210526_095054-1f77628b.pth'
cfg.work_dir = './tutorial_exps_faster_rcnn'
```

```python
cfg.optimizer.lr = 0.05
cfg.log_config.interval = 10
cfg.runner.max_epochs = 30
```

```python

cfg.evaluation.metric = 'bbox'

cfg.evaluation.interval = 12
cfg.checkpoint_config.interval = 12
cfg.lr_config.policy='step'

cfg.device = 'cuda'
cfg.seed = 0
set_random_seed(0, deterministic=False)
cfg.gpu_ids = range(1)
```

```python
from mmdet.datasets import build_dataset

datasets = [build_dataset(cfg.data.train)]
```

```python
from mmdet.datasets import build_dataloader
import torch
# 검증 데이터셋을 빌드합니다.
val_dataset = build_dataset(cfg.data.val)

# 검증 데이터 로더를 빌드합니다.
val_loader = build_dataloader(
    val_dataset,
    samples_per_gpu=1,  # 배치 크기 설정
    workers_per_gpu=1,
    dist=False,
    shuffle=False)

# 검증 데이터셋을 사용하여 모델을 평가합니다.

dataset = val_loader.dataset
model.cfg = cfg
det_results = []

for i, data in enumerate(dataset):
    with torch.no_grad():
        img = cv2.imread(data['img_metas'][0].data['filename'])
        result = inference_detector(model, img)
        det_results.append(result)
metrics = val_dataset.evaluate(det_results)
print(metrics)
```

```python
import matplotlib.pyplot as plt

# AR 값을 추출
ar_values_str = metrics['bbox_mAP_copypaste']
ar_values = [float(value) for value in ar_values_str.split()]

# IoU 임계값 리스트
iou_thresholds = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]

# AR 값이 10개씩 묶인 리스트로 변환
ar_values_split = [ar_values[i:i+10] for i in range(0, len(ar_values), 10)]

# 그래프 그리기
plt.figure(figsize=(10, 5))
for i, ar_values_iou in enumerate(ar_values_split):
    plt.plot(iou_thresholds[:len(ar_values_iou)], ar_values_iou, marker='o', linestyle='-', label=f'IoU={iou_thresholds[i]}')

plt.xlabel('IoU Threshold')
plt.ylabel('Average Recall (AR)')
plt.title('AR vs IoU Threshold')
plt.legend()
plt.grid(True)
plt.show()

```

```python
 import cv2

 # 입력 비디오 파일 경로와 출력 비디오 파일 경로 설정
 input_video_path = '/content/drive/MyDrive/대장용종 이미지 데이터/project/2.avi'
 output_video_path = '/content/drive/MyDrive/대장용종 이미지 데이터/project/12.avi'

 # 시작 시간과 종료 시간 설정 (단위: 초)
 start_time = 650  # 시작 시간 (예: 0초)
 end_time = 800 # 종료 시간 (예: 2분 = 120초)

 # 입력 비디오 파일 열기
 video_capture = cv2.VideoCapture(input_video_path)

 # 원래 비디오의 프레임 속도 및 프레임 크기 가져오기
 original_fps = int(video_capture.get(cv2.CAP_PROP_FPS))
 fps = original_fps / 2
 frame_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
 frame_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

 # 출력 비디오 파일 생성 (AVI 형식)
 fourcc = cv2.VideoWriter_fourcc(*'XVID')  # AVI 형식으로 설정
 output_video = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

 # 시작 시간까지 프레임 스킵
 video_capture.set(cv2.CAP_PROP_POS_FRAMES, int(start_time * fps))

 # 비디오를 시작 시간부터 읽어서 종료 시간까지의 프레임만 저장
 while True:
     ret, frame = video_capture.read()
     if not ret or video_capture.get(cv2.CAP_PROP_POS_FRAMES) > end_time * fps:
         break

     # 객체 탐지 수행
     result = inference_detector(model, frame)

     # 객체 탐지 결과를 프레임에 그리기
     frame = model.show_result(frame, result, score_thr=0.7)

     # 출력 비디오에 프레임 쓰기
     output_video.write(frame)

 # 파일 닫기
 video_capture.release()
 output_video.release()
```
